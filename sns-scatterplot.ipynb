```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib notebook
```

```python
df = pd.read_csv('Pokemon.csv')
grouped = df.groupby(df['Type 1']).count()
grouped.sort_values(by='Legendary').tail(7).index.values
df = df[df['Type 1'].isin(grouped.sort_values(by='Legendary').tail(7).index.values)]
df.head()
```

```python
sns.set(font_scale = 1.3)
g = sns.relplot(x='Attack', 
            y='Defense', 
            hue='Type 1', 
            size='Total',
            data=df,
            sizes=(40, 400), 
            alpha=.7, 
            palette='muted', 
            height=8, 
            aspect=8/8)

title = 'Properties of Pokemons by Type'
plt.title(title, fontsize=20, pad=20)
plt.subplots_adjust(top=0.85)

g._legend.texts[0].set_text('')
g._legend.texts[8].set_text('')
g._legend.set_bbox_to_anchor([1.01, .63])
plt.setp(g._legend.get_texts(), fontsize='16')
```

```python
filename = 'sns-scatterplot'
plt.savefig(filename+'.png')
```
